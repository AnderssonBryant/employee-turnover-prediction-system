```python
x = datasets.drop(['Attrition'],axis=1)
y = datasets.Attrition
```

```python
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
log_reg = LogisticRegression(max_iter= 100000)
log_reg.fit(x_train, y_train)
```

```python
from sklearn.metrics import (confusion_matrix,
                           accuracy_score,
                           precision_score,recall_score,f1_score)

y_pred = log_reg.predict(x_test)


# confusion matrix
cm = confusion_matrix(y_test, y_pred)
print ("Confusion Matrix : \n", cm)

# accuracy score of the model
print('Test accuracy = ', accuracy_score(y_test, y_pred))
print('Precision = ',precision_score(y_test, y_pred))
print('Recall = ',recall_score(y_test, y_pred))
print('F1 score = ',f1_score(y_test, y_pred))
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(x_train, y_train)
```

```python
y_resampled.value_counts()
```

```python
log_reg = LogisticRegression(max_iter= 100000)
log_reg.fit(X_resampled, y_resampled)
```

```python
y_pred = log_reg.predict(x_test)


# confusion matrix
cm = confusion_matrix(y_test, y_pred)
print ("Confusion Matrix : \n", cm)

# accuracy score of the model
print('Test accuracy = ', accuracy_score(y_test, y_pred))
print('Precision = ',precision_score(y_test, y_pred))
print('Recall = ',recall_score(y_test, y_pred))
print('F1 score = ',f1_score(y_test, y_pred))
```

```python
# Access the coefficients
coefficients = log_reg.coef_[0]

# Map coefficients to feature names
feature_names = list(x_train.columns)
feature_coefficients = dict(zip(feature_names, coefficients))

# Sort feature importance
sorted_feature_importance = sorted(feature_coefficients.items(), key=lambda x: abs(x[1]), reverse=True)

# Print feature importance
for feature, coef in sorted_feature_importance:
    print(f"{feature}: {coef}")
```

```python
prediction = log_reg.predict_proba(x)

positive_prediction = []
for item in prediction:
    formatted_num = f"{item[1]*100:.2f}"
    positive_prediction.append(formatted_num)

datasets['Results'] = positive_prediction

```
